### Basic line plots.

```python
plt.plot([1,2,3,4])
plt.xlabel("This is X-Label");
plt.ylabel("This is Y-Label");
plt.title('Our first Python plot')
plt.show()  
```

Note: x-axis ranges from 0-3 and y-axis from 1-4. If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0, 1, 2, 3].

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel("This is a plot")
plt.show() # this blocks execution.
```

### Scatter plots.

```python
plt.show(block=False) # non-blocking way with Matplotlib
plt.plot([1,2,3,4], [1,4,9,16], "ro")
plt.axis([0, 6, 0, 20])
plt.show()
```

Experment: To build a vector of 10,000 normal deviates with variance = 0.5^2, and mean = 2.

```python
mu, sigma = 2, 0.5
data = np.random.normal(mu,sigma,10000)
# Plot a normalized histogram with 50 bins
plt.hist(data, bins=50, density=1)       # matplotlib version (plot)
plt.ylabel("Probability")
plt.show()
```

### Subplots.

```python
def f(x):
    return np.exp(-x) * np.cos(2*np.pi*x)

def g(x):
    return np.sin(2*np.pi*x) / x

x1 = np.arange(0.001, 2*np.pi, 0.25)
x2 = np.arange(0.001, 2*np.pi, 0.02)

# subplot(nrows, ncols, index, **kwargs)
plt.subplot(221)
plt.plot(x1, f(x1), "bo", x2, f(x2), "k") ## k means black line

plt.subplot(222)
plt.plot(x2, np.cos(x2), "r--")

plt.subplot(223)
plt.plot(x1, g(x1), "mo", x2, g(x2), "k")

plt.subplot(224)
plt.plot(x2, np.sin(x2), "g--")

plt.show()
```

### Histograms.

alpha value: Setting the opacity.

plt.text: Add the text s to the axes at location x, y in data coordinates.

The r preceding the title string is important -- it signifies that the string is a raw string and not to treat backslashes as python escapes.

```python
mu = 100
sigma = 15
x = mu + sigma*np.random.randn(10000);
num_bins = 20; 
plt.hist(x, num_bins, facecolor="orange", density=True, alpha=0.75);

plt.title("Histogram of IQ")
plt.xlabel("Smarts", fontsize=18, color='red')
plt.ylabel("Probability")
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')  # Latex for math typing

plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
